```python
import billboard
import numpy as np
import pandas as pd
import sys
import os
import time
import datetime as dt
```

### Using billboard.py to scrape billboard chart
Repo found [here](https://github.com/guoguo12/billboard-charts)
* Date Monday of the week
* Artsit: artist name
* isNew: if this song is a new song
* lastPos: the rank of last wekk
* peakPos: the peak of rank
* title: song title
* weeks: how long does it stay on billboard

```python
# all constant here
begin = dt.datetime.strptime('2018-11-10','%Y-%m-%d')
date_format = '%Y-%m-%d'
chart_name = 'r-b-hip-hop-songs'
search_date = begin
df = {i:[] for i in ['Date','artist','isNew','lastPos','peakPos','rank','title','weeks']}
```

```python
# Download data week by week.
while(search_date.year > 1958):
    print('Searching {}'.format(dt.datetime.strftime(search_date,
                                         date_format)))
    chart = billboard\
    .ChartData(chart_name,
               date=dt.datetime.strftime(search_date,
                                         date_format))
    for each in chart:
        df['Date'].append(dt.datetime.strftime(search_date,date_format))
        df['artist'].append(each.artist)
        df['isNew'].append(each.isNew)
        df['lastPos'].append(each.lastPos)
        df['peakPos'].append(each.peakPos)
        df['rank'].append(each.rank)
        df['title'].append(each.title)
        df['weeks'].append(each.weeks)
    search_date = search_date - dt.timedelta(days = 7)
```

```python
final_df = pd.DataFrame(df)
```
