```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import multiprocessing
import csv
```

```python
tracks = pd.read_csv("spotify_data/tracks.csv")
tracks_name = list(set(tracks['name']))
base_url = "http://api.genius.com"
headers = {
    "Authorization":'Bearer sn-ocjkB-V37CZmM1_7mUcYHETZWH2ukWm46iRnERc6UcRr99GdexFezZBaQsFiL'
}

def search_song(i):
    tmp = pd.DataFrame(columns=["track_name", "artist_name", "lyrics"])
    song_title = tracks_name[i]
    print(str(i)+':'+song_title)
    search_url = base_url + "/search?q={}".format('%20'.join(song_title.split(' ')))
    response = requests.get(search_url, headers=headers)
    json = response.json()
    for hit in range(len(json['response']['hits'])):
        lyrics_url = json['response']['hits'][hit]['result']['url']
        artist_name = json['response']['hits'][hit]['result']['primary_artist']['name']
        page = requests.get(lyrics_url)
        if page.status_code == 404:
            tmp = tmp.append({'track_name':song_title, 'artist_name':None, 'lyrics':None},ignore_index=True)
        else:
            html = BeautifulSoup(page.text,'html.parser')
            l = html.find('div',class_= 'lyrics')
            if l == None:
                tmp = tmp.append({'track_name':song_title, 'artist_name':None, 'lyrics':None},ignore_index=True)
            else:    
                lyrics = l.get_text()
                tmp = tmp.append({'track_name':song_title, 'artist_name':artist_name, 'lyrics':lyrics},ignore_index=True)
    return tmp     
```

```python
result = []
pool = multiprocessing.Pool(multiprocessing.cpu_count())
for i in range(20000,30000):
    p = pool.apply_async(search_song,args=(i,))
    print(p)
    result.append(p)
pool.close()
pool.join()
```
